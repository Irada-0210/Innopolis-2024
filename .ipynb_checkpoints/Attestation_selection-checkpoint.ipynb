```python
import pandas as pd
datasets = pd.read_csv("https://raw.githubusercontent.com/vincentarelbundock/Rdatasets/master/datasets.csv")
datasets.head()
```

#### Выберем те, у которых достаточно наблюдений

```python
datasets = datasets.loc[(datasets['Rows']>500)&(datasets['Cols']>5)&(datasets['n_factor']>2)&(datasets['n_numeric']>2)]
datasets.head()
```

```python
items = [
    'CollegeDistance', 'CreditCard', 'DoctorVisits', 'Guns', 'HMDA', 'TravelMode', 'nassCDS', 'Benefits', 'OFP', 
    'VietNamH', 'diamonds', 'credit_data', 'lending_club', 'mlc_churn', 'wa_churn', 'email', 'happiness', 'InstInnovation',
    'HousePrices', 'Chile', 'Mroz', 'Computers', 'Griliches', 'Aids2', 'birds'
]

len(items)
```

```python
datasets[datasets['Item'].isin(items)].to_csv(r'C:\Courses\Innopolis\2024\ПИШ\Домашние задания\Аттестация2\datasets.csv')
```
